```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating Dataset 1 of Anscombe's Quartet
data = {
    'x': [10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0],
    'y': [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]
}


# Convert to Pandas DataFrame
df = pd.DataFrame(data)

# Compute Basic Statistics
print("Dataset 1 - Basic Statistics")
print(df.describe())

# Scatter Plot
plt.figure(figsize=(6,4))
plt.scatter(df['x'], df['y'], color='blue', marker='D', label='Data Points')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title("Anscombe's Quartet - Dataset 1")
#plt.axhline(df['y'].mean(), color='red', linestyle='dashed', label='Mean of Y')
#plt.axvline(df['x'].mean(), color='green', linestyle='dashed', label='Mean of X')

x_min, x_max = min(df['x']), max(df['x'])  # Get min and max X values
plt.plot([x_min, x_max], [x_min, x_max], color='grey', linestyle='dashed', label='45-degree Line (y=x)')

#plt.legend()
plt.grid(True)
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Creating sample data
x = np.linspace(1, 10, 10)  # X values from 1 to 10
y1 = x * 1.5  # First trend
y2 = x ** 1.2  # Second trend
y3 = np.log(x) * 5  # Third trend (logarithmic)
y4 = np.sqrt(x) * 3  # Fourth trend (square root function)

# Plotting the four lines with different colors
plt.plot(x, y1, label='Linear Growth', color='red', linestyle='-', marker='o')
plt.plot(x, y2, label='Power Growth', color='blue', linestyle='--', marker='s')
plt.plot(x, y3, label='Logarithmic Growth', color='green', linestyle='-.', marker='^')
plt.plot(x, y4, label='Square Root Growth', color='purple', linestyle=':', marker='d')

# Adding titles and labels
plt.xlabel('X-axis (Time or Variable)')
plt.ylabel('Y-axis (Values)')
plt.title('Line Chart with Different Colored Lines')
plt.legend()  # Show legend to differentiate lines

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating synthetic dataset (Height vs. Body Mass)
np.random.seed(42)  # For reproducibility
dataSize = 1000

height = np.random.normal(loc=170, scale=10, size=dataSize)  # Heights in cm (mean=170, std=10)
weight = height * 0.5 + np.random.normal(loc=0, scale=5, size=dataSize)  # Weight depends on height

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(height, weight, color='blue', alpha=0.5, edgecolors='black')

# Labels and title
plt.xlabel("Height (cm)")
plt.ylabel("Body Mass (kg)")
plt.title("Scatter Plot: Height vs. Body Mass")
plt.grid(True)

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Sample Data: Cities, Population, Temperature, and Pollution Index
data = {
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix", "Philadelphia", "San Antonio", "San Diego"],
    "Population": [8419600, 3980400, 2716000, 2328000, 1690000, 1584200, 1543200, 1424100],  # Inhabitants
    "Avg_Temperature": [13.3, 18.7, 11.6, 21.0, 24.6, 12.8, 22.4, 17.5],  # Celsius
    "Pollution_Index": [60, 75, 55, 70, 50, 65, 45, 40]  # Higher means more pollution
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Normalize Bubble Size (Population) and Color (Pollution Index)
bubble_size = np.log(df["Population"]) * 300  # Log scaling for better visualization
colors = df["Pollution_Index"]

# Create Bubble Chart
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df["Avg_Temperature"], df["Pollution_Index"], 
                       s=bubble_size, c=colors, cmap="coolwarm", alpha=0.6, edgecolors="black")

# Add Annotations (City Names)
for i, txt in enumerate(df["City"]):
    plt.annotate(txt, (df["Avg_Temperature"][i], df["Pollution_Index"][i]), fontsize=10, ha="right", va="bottom")

# Customizations
plt.colorbar(scatter, label="Pollution Index")  # Color scale legend
plt.xlabel("Average Temperature (C)", fontsize=12)
plt.ylabel("Pollution Index", fontsize=12)
plt.title("Bubble Chart: City Population, Temperature & Pollution Index", fontsize=14)
plt.xticks(np.arange(10, 30, 2))  # Custom X-axis ticks
plt.grid(True, linestyle="--", alpha=0.5)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data: Different categories and their values
categories = ["Electronics", "Clothing", "Groceries", "Furniture", "Toys"]
values = [450, 300, 700, 200, 150]  # Sales in USD

# --- Bar Plot ---
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)  # First subplot for Bar Chart
plt.bar(categories, values, color=['blue', 'green', 'red', 'orange', 'purple'])
plt.xlabel("Categories")
plt.ylabel("Sales (USD)")
plt.title("Bar Chart: Sales by Category")
plt.xticks(rotation=20)  # Rotates labels for better readability
plt.grid(axis="y", linestyle="--", alpha=0.5)  # Light grid lines for readability

# --- Pie Chart ---
plt.subplot(1, 2, 2)  # Second subplot for Pie Chart
colors = ['blue', 'green', 'red', 'orange', 'purple']
explode = [0.1, 0, 0, 0, 0]  # Slightly explode the first slice for emphasis

plt.pie(values, labels=categories, autopct="%1.1f%%", colors=colors, explode=explode,
        startangle=140, wedgeprops={'edgecolor': 'black'})

plt.title("Pie Chart: Sales Distribution")

# Show the plots
plt.tight_layout()  # Adjust spacing to fit both plots
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data: Test Scores of Students in Three Classes
np.random.seed(42)  # For reproducibility
class_A = np.random.randint(50, 100, 20)  # Scores between 50 and 100
class_B = np.random.randint(60, 95, 20)   # Scores between 60 and 95
class_C = np.random.randint(40, 90, 20)   # Scores between 40 and 90

# Combine data into a list
data = [class_A, class_B, class_C]

# Create Box Plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=["Class A", "Class B", "Class C"], patch_artist=True)

# Customizing the Plot
plt.title("Box Plot of Student Test Scores")
plt.xlabel("Classes")
plt.ylabel("Test Scores")
plt.grid(axis="y", linestyle="--", alpha=0.5)  # Adding a light grid for better readability

# Show the plot
plt.show()
```
