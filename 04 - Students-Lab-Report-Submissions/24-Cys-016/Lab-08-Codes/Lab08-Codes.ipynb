```python
# Alpha-Beta Pruning Implementation

import math

def alphabeta(depth, node, maximizing, values, alpha, beta):
    # If we reached terminal node
    if depth == 3:
        return values[node]

    if maximizing:
        best = -math.inf
        for child in tree[node]:
            value = alphabeta(depth + 1, child, False, values, alpha, beta)
            best = max(best, value)
            alpha = max(alpha, best)
            if beta <= alpha:
                print("Pruned at node:", child)
                break
        return best

    else:
        best = math.inf
        for child in tree[node]:
            value = alphabeta(depth + 1, child, True, values, alpha, beta)
            best = min(best, value)
            beta = min(beta, best)
            if beta <= alpha:
                print("Pruned at node:", child)
                break
        return best


# Tree structure from the picture
tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': ['H', 'I'],
    'E': ['J', 'K'],
    'F': ['L', 'M'],
    'G': ['N', 'O']
}

# Terminal node values
values = {
    'H': 4,
    'I': 2,
    'J': -3,
    'K': -6,
    'L': 7,
    'M': 0,
    'N': 5,
    'O': 8
}

# Run Alpha-Beta Pruning
result = alphabeta(0, 'A', True, values, -math.inf, math.inf)
print("\nOptimal value at root (A):", result)
```

```python
# Simple Minimax for Tic-Tac-Toe

import math

# Check winner
def check_winner(board):
    win_positions = [
        (0,1,2),(3,4,5),(6,7,8),
        (0,3,6),(1,4,7),(2,5,8),
        (0,4,8),(2,4,6)
    ]
    for a,b,c in win_positions:
        if board[a] == board[b] == board[c] and board[a] != " ":
            return board[a]
    if " " not in board:
        return "Draw"
    return None

# Minimax function
def minimax(board, is_max):
    winner = check_winner(board)
    if winner == "X":
        return 1
    if winner == "O":
        return -1
    if winner == "Draw":
        return 0

    if is_max:
        best = -math.inf
        for i in range(9):
            if board[i] == " ":
                board[i] = "X"
                best = max(best, minimax(board, False))
                board[i] = " "
        return best
    else:
        best = math.inf
        for i in range(9):
            if board[i] == " ":
                board[i] = "O"
                best = min(best, minimax(board, True))
                board[i] = " "
        return best

# Given board from picture
board = [
    "O","X","O",
    "O","X","X",
    "X","O"," "
]

best_move = -1
best_score = -math.inf

for i in range(9):
    if board[i] == " ":
        board[i] = "X"
        score = minimax(board, False)
        board[i] = " "
        if score > best_score:
            best_score = score
            best_move = i

print("Best move for X is position:", best_move)
```
