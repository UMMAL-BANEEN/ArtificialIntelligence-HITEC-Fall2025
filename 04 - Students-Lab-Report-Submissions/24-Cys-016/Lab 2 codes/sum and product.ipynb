#  Sample Sum and Product of Numbers in a Tuple Code

```python

numbers = (1, 2, 4, 2, 6)

sum_result = sum(numbers)

product_result = 1
for n in numbers:
    product_result = product_result * n

print("Sum is:", sum_result)
print("Product is:", product_result)
```

# Code With Different Numbers

```python

numbers = (3, 5, 7, 2, 8)

sum_result = sum(numbers)

product_result = 1
for n in numbers:
    product_result = product_result * n

print("Sum is:", sum_result)
print("Product is:", product_result)
```
