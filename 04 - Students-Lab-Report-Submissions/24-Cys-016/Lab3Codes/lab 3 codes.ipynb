```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Example:
num = 6
print("The factorial of", num, "is:", factorial(num))
```

```python
def unique_list(lst):
    unique = []
    for item in lst:
        if item not in unique:
            unique.append(item)
    return unique

sample_list = [1, 2, 3, 3, 4, 4, 5, 3, 7, 5]
print("Unique List:", unique_list(sample_list))
```

```python
def show_even_numbers(n):
    even_numbers = []
    for i in range(2, 2*n + 1, 2):
        even_numbers.append(i)
    print("First", n, "even numbers are:", even_numbers)
    return even_numbers
# Example: calling the function
n = 6
evens = show_even_numbers(n)

# Calculate sum and product outside the function
sum_even = sum(evens)

product_even = 1
for num in evens:
    product_even *= num
print("Sum of even numbers:", sum_even)
print("Product of even numbers:", product_even)
```

```python
def match_strings(str1, str2):
    if str1 == str2:
        print("Both strings are the same.")
    else:
        print("Strings are different.")

# Example:
string1 = input("Enter first string: ")
string2 = input("Enter second string: ")

match_strings(string1, string2)
```
