```python
# Person class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
# Job class
class Job:
    def __init__(self, designation, salary):
        self.designation = designation
        self.salary = salary
# Employee class inheriting both Person and Job
class Employee(Person, Job):
    def __init__(self, name, age, designation, salary):
        Person.__init__(self, name, age)
        Job.__init__(self, designation, salary)
    def display_details(self):
        print("Employee Details:")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Designation:", self.designation)
        print("Salary:", self.salary)

# Example:
emp1 = Employee("Fatima", 19, "Digital Forensics Expert", 80000)
emp1.display_details()
```

```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius

# Example:
r = float(input("Enter the radius of the circle: "))
circle1 = Circle(r)

print("Area of the circle:", circle1.area())
print("Perimeter of the circle:", circle1.perimeter())
```

```python
class Power:
    def pow(self, x, n):
        result = 1
        if n > 0:
            for i in range(n):
                result *= x
        elif n < 0:
            for i in range(-n):
                result *= x
            result = 1 / result
        else:
            result = 1
        return result
# Example:
p = Power()
x = float(input("Enter base number (x): "))
n = int(input("Enter power (n): "))

print(f"{x} raised to the power {n} is:", p.pow(x, n))
```

```python
# Base class
class Item:
    def __init__(self, name, description, price):
        self.name = name
        self.description = description
        self.price = price

    def viewFullDescription(self):
        print(f"Item: {self.name}")
        print(f"Description: {self.description}")
        print(f"Price: {self.price}")

    def addToShoppingBasket(self):
        print(f"{self.name} added to shopping basket.")

    def removeFromShoppingBasket(self):
        print(f"{self.name} removed from shopping basket.")


# MP3 subclass
class MP3(Item):
    def __init__(self, name, description, price, artist, duration):
        super().__init__(name, description, price)
        self.artist = artist
        self.duration = duration

    def play(self):
        print(f"Playing MP3 by {self.artist}...")

    def download(self):
        print(f"Downloading {self.name}...")


# DVD subclass
class DVD(Item):
    def __init__(self, name, description, price, certificate, duration, actors):
        super().__init__(name, description, price)
        self.certificate = certificate
        self.duration = duration
        self.actors = actors

    def viewTrailer(self):
        print(f"Viewing trailer for {self.name}...")


# Book subclass
class Book(Item):
    def __init__(self, name, description, price, author, numberOfPages, genre):
        super().__init__(name, description, price)
        self.author = author
        self.numberOfPages = numberOfPages
        self.genre = genre

    def previewContent(self):
        print(f"Previewing content of the book '{self.name}' by {self.author}...")


# Example usage:
mp3 = MP3("Song A", "A popular song", 150, "Atif Aslam", "4:30")
dvd = DVD("Movie B", "Action movie", 500, "PG-13", "2h", ["Actor1", "Actor2"])
book = Book("Python Basics", "Learn Python easily", 800, "Ali Khan", 250, "Education")

mp3.viewFullDescription()
mp3.play()

dvd.addToShoppingBasket()
dvd.viewTrailer()

book.previewContent()
```
