# Number in Word Form Code

```python

numbers_dict = {
    '0': 'Zero',
    '1': 'One',
    '2': 'Two',
    '3': 'Three',
    '4': 'Four',
    '5': 'Five',
    '6': 'Six',
    '7': 'Seven',
    '8': 'Eight',
    '9': 'Nine'
}

num = input("Enter a number: ")

for digit in num:
    print(numbers_dict[digit], end=" ")
```
