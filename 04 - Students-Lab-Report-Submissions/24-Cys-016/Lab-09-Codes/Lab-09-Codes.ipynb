```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (file name apni CSV ka likho)
df = pd.read_csv("heart_disease.csv")

print(df.head())
```

```python
mean_age = df['Age'].mean()
print("Mean Age:", mean_age)
```

```python
print(df['Heart Disease Status'].value_counts())


```

```python
print("Max Cholesterol:", df['Cholesterol Level'].max())
print("Min Cholesterol:", df['Cholesterol Level'].min())

```

```python
print("BMI over 30:", df[df['BMI'] > 30].shape[0])
```

```python
count = df[
    (df['High Blood Pressure'] == 'Yes') &
    (df['Heart Disease Status'] == 'Yes')
].shape[0]

print("High BP & Heart Disease:", count)
```

```python
plt.hist(df['BMI'], bins=10)
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.title("BMI Distribution")
plt.show()
```

```python
exercise_hd = df.groupby('Exercise Habits')['Heart Disease Status'].value_counts().unstack()

exercise_hd.plot(kind='bar')
plt.xlabel("Exercise Habits")
plt.ylabel("Count")
plt.title("Heart Disease vs Exercise Habits")
plt.show()
```

```python
risk = df.groupby('Exercise Habits')['Heart Disease Status'].value_counts(normalize=True)
print(risk)
```

```python
smoking_counts = df['Smoking'].value_counts()

plt.pie(smoking_counts, labels=smoking_counts.index, autopct='%1.1f%%')
plt.title("Smokers vs Non-Smokers")
plt.show()
```

```python
percentage_smokers = (smoking_counts['Yes'] / len(df)) * 100
print("Percentage of Smokers:", percentage_smokers)
```

```python
plt.scatter(df['BMI'], df['Triglyceride Level'])
plt.xlabel("BMI")
plt.ylabel("Triglyceride Level")
plt.title("BMI vs Triglyceride Level")
plt.show()
```

```python
plt.boxplot(df['Sleep Hours'])
plt.ylabel("Sleep Hours")
plt.title("Sleep Hours Distribution")
plt.show()
```

```python
print("Median Sleep Hours:", df['Sleep Hours'].median())
```
