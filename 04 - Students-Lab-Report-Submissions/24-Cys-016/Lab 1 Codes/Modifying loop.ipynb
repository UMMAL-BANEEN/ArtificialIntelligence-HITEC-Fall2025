# Modifying Loop code

```python
# Part 1: Print numbers from 1 to 10 using while loop
print("Numbers from 1 to 10:")
i = 1
while i <= 10:
    print(i)
    i = i + 1

# Part 2: Print even numbers between 1 and 20
print("\nEven numbers between 1 and 20:")
i = 2
while i <= 20:
    print(i)
    i = i + 2
```
